```python
import requests
import json
from bs4 import BeautifulSoup

# Function to scrape and store quotes
def scrape_and_store(url, storage_file):
    try:
        # Send an HTTP GET request to the URL
        response = requests.get(url)
        
        # Check if the request was successful (status code 200)
        if response.status_code == 200:
            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(response.text, 'html.parser')
            print(response.text)
            # Extract specific content (in this case, quotes and authors)
            quotes = []
            for quote in soup.find_all('div', class_='quote'):
                text = quote.find('span', class_='text').get_text().strip()  # Get the quote text
                author = quote.find('small', class_='author').get_text().strip()  # Get the author
                quotes.append({"text": text, "author": author})

            # Store the scraped quotes with other relevant data
            data_to_store = {
                "title": soup.title.string.strip() if soup.title else "No title",
                "url": url,
                "quotes": quotes  # Store all the quotes found
            }
            
            # Open the storage file in append mode and store the data
            with open(storage_file, 'a') as f:
                json.dump(data_to_store, f, indent=4)
                f.write('\n')  # Write a new line for separation between records
            
            print(f"Data from {url} stored successfully.")
        else:
            print(f"Failed to retrieve data from {url}. Status code: {response.status_code}")
    
    except Exception as e:
        print(f"An error occurred: {str(e)}")

# Example usage
start_url = 'http://quotes.toscrape.com/'
storage_file = 'scraped_data.json'

# Call the function to scrape and store data
scrape_and_store(start_url, storage_file)
```
